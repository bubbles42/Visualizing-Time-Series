```python
import pandas as pd

df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv")
df.info()
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df.head()
```

```python
df.info()
```

```python
df.set_index('date', inplace=True)
df.head()
```

```python
df = df.loc['2000':, ['precipitation','mean_temp','min_temp','max_temp','snow_depth']]
df.head()
```

```python
df.isna().sum()
```

```python
pd.set_option('display.max_rows', 1000)
df.head(500)
```

```python
df['mean_temp'] = df['mean_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()

df.head(500)
```

```python
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
df.isna().sum()
```

```python
from matplotlib import dates as mdates
```

```python
tempdf = df.loc['2018':,['mean_temp','min_temp','max_temp']]
tempdf.head()
```

```python
ax = tempdf.plot(figsize=(12,3))
ax.set(title='Daily Temperature', ylabel='Degrees')

majorLoc = mdates.YearLocator(1)
majorFmt = mdates.DateFormatter('%Y')

minorLoc = mdates.MonthLocator(interval=3)
minorFmt = mdates.DateFormatter('%b')

ax.xaxis.set_major_locator(majorLoc)
ax.xaxis.set_major_formatter(majorFmt)
ax.xaxis.set_minor_locator(minorLoc)
ax.xaxis.set_minor_formatter(minorFmt)

fig=ax.get_figure()
fig.autofmt_xdate(which='major', rotation=90);
```
